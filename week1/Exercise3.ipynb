```python
import pandas as pd
import sqlite3

db = sqlite3.connect("history_of_baseball.sqlite")


```

### Exercise 3.3: Join together two tables

```python
hofquery = """
    SELECT *
    FROM player AS p 
    INNER JOIN hall_of_fame AS hof 
    ON p.player_id = hof.player_id 
    WHERE hof.inducted='Y';
"""

hof_results = pd.read_sql(hofquery, db)
print(hof_results)
```

### Exercise 3.4: Join together three tables (oh my!)

```python
collegequery = """
    SELECT DISTINCT pc.player_id, pc.college_id 
    FROM hall_of_fame AS hof 
    INNER JOIN player AS p
    on p.player_id = hof.player_id
    INNER JOIN player_college AS pc 
    ON hof.player_id = pc.player_id 
    WHERE hof.inducted='Y'
    ORDER BY pc.college_id;
"""

college_results = pd.read_sql(collegequery, db).groupby('college_id').count()
print(college_results)
```
